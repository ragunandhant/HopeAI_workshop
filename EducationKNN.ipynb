```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/College_placement (1).csv")
```

```python
df.info()
```

```python
df
```

```python
df = pd.get_dummies(df,drop_first=True)
```

```python
df
```

```python
independent = df.drop(["PlacedOrNot"],axis=1)
dependent = df["PlacedOrNot"]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(independent,dependent,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
model.fit(xtrain,ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(ytest, ypred)
```

```python
print(clf_report)
```

```python
param_grid = {
    'n_neighbors':[3,10,4,5,6,7,8,9,10,11,12],
    'weights':['uniform', 'distance'],
    'algorithm':['auto', 'ball_tree', 'kd_tree', 'brute'],
    'leaf_size':[100,50,60,40,30,70,80,150,120,130,200]
}
```

```python
from sklearn.model_selection import GridSearchCV

best = GridSearchCV(KNeighborsClassifier(),param_grid,cv=5,refit=True,scoring = 'f1')
best.fit(xtrain,ytrain)
best.best_estimator_
```

```python
ypred = best.predict(xtest)
from sklearn.metrics import classification_report
clf_report = classification_report(ytest, ypred)
print(clf_report)
```
